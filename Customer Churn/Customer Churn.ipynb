```python
df = pd.read_csv(file_path)
```

```python
df.drop(["customerID"], axis = 1, inplace = True)
```

```python
sns.countplot(x = "Churn", data = df)
plt.title("churn dist")
plt.show()
```

```python
new_df = pd.get_dummies(df, drop_first = True)
```

```python
new_df
```

```python
from sklearn.model_selection import train_test_split as TTS
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
X = new_df.drop("Churn_Yes", axis = 1)
Y = new_df["Churn_Yes"]
```

```python
train_data, test_data, train_labels, test_labels = TTS(X, Y, test_size = 0.2, shuffle = True)
```

```python
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.fit_transform(test_data)
```

```python
model = LogisticRegression()
model.fit(train_data, train_labels)
```

```python
labels_pred = model.predict(test_data)
```

```python
print("Accuracy", accuracy_score(test_labels, labels_pred))
```

```python
classification_report(test_labels, labels_pred)
```

```python
import joblib
```

```python
joblib.dump(model, "Churn_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```
