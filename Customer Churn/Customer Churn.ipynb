```python
X = df.drop("Churn", axis = 1)
Y = df["Churn"]

df_minority = df[df["Churn"] == "Yes"]
df_majority = df[df["Churn"] == "No"]

```

```python
from sklearn.utils import resample
```

```python
df_majority_downsampled = resample(df_majority, 
                                   replace = False,
                                   n_samples = len(df_minority),
                                   random_state = 42
                                  )
df_balanced = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_balanced.head()
```

```python
sns.countplot(x = "Churn", data = df_balanced)
plt.title("churn dist")
plt.show()
```

```python
df_balanced.isna().sum()
```

```python
df_balanced.drop('TotalCharges', axis = 1, inplace = True)
```

```python
encoders = {}

for col in df_balanced.select_dtypes(include = ["object"]).columns:
    enc = LabelEncoder()
    df_balanced[col] = enc.fit_transform(df_balanced[col])
    encoders[col] = enc
```

```python
new_data = df_balanced.drop("Churn", axis = 1)
new_labels = df_balanced["Churn"]
```

```python
from sklearn.model_selection import train_test_split as TTS
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import joblib
```

```python
train_data, test_data, train_labels, test_labels = TTS(new_data, new_labels, train_size = 0.8, shuffle = True)
```

```python
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.fit_transform(test_data)
```

```python
model = LogisticRegression()
model.fit(train_data, train_labels)

pred_labels = model.predict(test_data)
```

```python
print("acc", accuracy_score(test_labels, pred_labels))
```

```python
classification_report(test_labels, pred_labels)
```
